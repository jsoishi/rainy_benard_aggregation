```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import erf
```

```python
plt.style.use("prl")
```

```python
def H(x, k):
    return 0.5*(1+erf(k*x))
```

```python
q_m_qs = np.linspace(-0.001,0,1000)
```

```python
plt.plot(q_m_qs, H(q_m_qs,1e3),label='k=1e3')
plt.plot(q_m_qs, H(q_m_qs,1e4),label='k=1e4')
plt.plot(q_m_qs, H(q_m_qs,1e5),label='k=1e5')
plt.plot(q_m_qs, H(q_m_qs,1e6),label='k=1e6')
```

```python
def N(x,k):
    return H(x,k) + k*x/np.sqrt(np.pi)*np.exp(-k**2 * x**2)
```

```python
for k in np.logspace(3,6,5,endpoint=False):
    plt.plot(q_m_qs,N(q_m_qs,k),label=f"k = {k:.1e}")
k = 1e6
plt.plot(q_m_qs,N(q_m_qs,k),label=f"k = {k:.1e}",color='black', lw=2)
plt.legend()
plt.xlabel(r"$q-q_s$")
plt.ylabel(r"$\mathcal{N}$")
```

```python
plt.plot(q_m_qs,N(q_m_qs,1e4),label=r"$\mathcal{N}$")
plt.plot(q_m_qs, H(q_m_qs,1e4),label=r"$\mathcal{H}$")
plt.xlabel(r"$q-q_s$")
plt.ylabel(r"$\mathcal{N},\mathcal{H}$")
plt.legend()
```
