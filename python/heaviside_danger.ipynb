```python
from scipy.special import erf
```

```python
k = 10

x = np.linspace(-1,1,1000)
ef = 0.5*(1 + erf(k*x))
th = 0.5*(np.tanh(k*x))
heaviside = np.piecewise(x, [x < 0, x >= 0], [0, 1])

plt.plot(x,ef)
plt.plot(x,heaviside,color='k',alpha=0.4)
plt.fill_between(x,heaviside,ef,where=x<0,color='red',alpha=0.4)
plt.xlabel('$q-q_s$')
plt.ylabel('$\mathcal{H}(q-q_s)$')
plt.annotate(r"$C = \frac{(q-q_s)}{\tau} \mathcal{H}(q-q_s) < 0$",(-0.05,0.05),(-0.9,0.5),arrowprops = dict(
    arrowstyle="->",
    ),fontsize=14)
plt.xlim(-1,1)
plt.tight_layout()
plt.savefig('heaviside_danger.pdf')
```
